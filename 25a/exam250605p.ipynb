### 1. Resuelve, usando el método de bisección, la ecuación $f(x) = x^3 - 4x + 1$ en el intervalo $[0, 2]$. Realiza tres iteraciones.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 4*x + 1;

x = np.linspace(-1, 3)
y = f(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label=r'$f(x) = x^3 -4x +1$')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.title(r'Gráfica de $f(x) = x^3 -4x +1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

a = 0;
b = 2;

print(f'{"i":>2} {"a":>10} {"b":>10} {"pm":>10}')
for i in range(3):
    pm = (a + b) / 2
    print(f'{i:2d} {a:10.4f} {b:10.4f} {pm:10.4f}')

    if f(a) * f(pm) < 0 :
        b = pm
    else:
        a = pm

print (f'\nSolución aproximada en: {pm:10.4f}')
```

### 2. Usa el método de Newton-Raphson para encontrar una raíz de la ecuación $f(x) = \cos(x) - x$ con $x_0 = 0.5$. Realiza dos iteraciones.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x) - x

def df(x):
    return -np.sin(x) - 1
    
x = np.linspace(-2, 2)
y = f(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label=r'$f(x) = \cos(x) - x$')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.title(r'Gráfica de $f(x) = \cos(x) - x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()


x = 0.5 # propuesta del valor inicial

print(f'{"i":>2} {"x_i":>10} {"f(x_i)":>10} {"f\'(x_i)":>10} {"x_{i+1}":>10}')
for i in range(2):
    fx = f(x)
    dfx = df(x)
    x_new = x - fx / dfx
    print(f'{i:2d} {x:10.4f} {fx:10.4f} {dfx:10.4f} {x_new:10.4f}')
    x = x_new

print (f'\nSolución aproximada en: {x_new:10.4f}')
```

### 4. Plantea la ecuación de punto fijo equivalente para la ecuación $e^{-x} = x$. ¿Cuál sería la función $g(x)$ y cómo determinarías si el método converge?

Ecuación del punto fijo:
$$x=g(x)$$

La ecuación es:
$$e^{−x}=x$$

Se puede escribir como:
$$x=e^{−x}$$

La función de iteración de punto fijo:
$$g(x)=e^{−x}$$

El método de punto fijo converge si, en un intervalo alrededor de la raíz, se cumple que:
$$∣g′(x)∣<1$$

### 6. Resuelve el siguiente sistema de ecuaciones usando una sola iteración del método de Gauss-Seidel. Usa $x_1^{(0)} = 0$, $x_2^{(0)} = 0$:
$$\begin{cases}
3x_1 + x_2 = 9 \\
x_1 + 4x_2 = 8
\end{cases}$$

Procedimiento para Gauss-Seidel:}

**Paso 1:** Despeja cada incógnita:

$$x_1 = \frac{9 - x_2}{3}$$

$$x_2 = \frac{8 - x_1}{4}$$

**Paso 2:** Valores iniciales $x_1^{(0)} = 0, x_2^{(0)} = 0$:

$$x_1^{(1)} = \frac{9 - x_2^{(0)}}{3} = \frac{9 - 0}{3} = 3$$

$$x_2^{(1)} = \frac{8 - x_1^{(1)}}{4} = \frac{8 - 3}{4} = \frac{5}{4} = 1.25$$

Después de una iteración,
$$x_1^{(1)} = 3, \quad x_2^{(1)} = 1.25$$

### 7. Explica brevemente la diferencia entre los métodos de Gauss y Gauss-Jordan para la solución de sistemas lineales.

El método de Gauss (también llamado eliminación de Gauss) es un procedimiento para resolver sistemas de ecuaciones lineales. Consiste en transformar la matriz del sistema en una forma triangular superior, lo que permite luego resolver el sistema por sustitución hacia atrás.

El método de Gauss-Jordan es una extensión del método de Gauss. Se continúa el proceso de eliminación hasta obtener una matriz diagonal o forma reducida por filas (es decir, unos en la diagonal y ceros en el resto de la columna). De esta forma, cada incógnita se puede leer directamente del sistema, sin necesidad de hacer sustitución hacia atrás.

### 10. Utiliza el polinomio interpolación de Lagrange para interpolar el valor de $f(2)$ dados los puntos $(1, 1)$, $(3, 3)$, $(4, 13)$

```python
import numpy as np

x = np.array([1, 3, 4])
y = np.array([1, 3, 13])

# Valor a buscar
xx = 2.0

n = len(x)

suma = 0.0
for i in range(n):
    multi = 1.0
    for j in range(n):
        if i != j:
            multi *= (xx - x[j]) / (x[i] - x[j])
    suma += multi * y[i]

print(f'Aproximación: {suma:.4f}')
```

### 11. Explica la diferencia entre el método del trapecio y el método de Simpson 1/3 para integración numérica.

Método del trapecio aproxima la curva entre cada par de puntos usando una línea recta. Es decir, divide el área bajo la curva en varios trapecios (de ahí su nombre), y suma sus áreas. Es fácil de aplicar y da buenos resultados si la función es aproximadamente lineal en cada subintervalo, pero puede ser menos precisa si la función es muy curva.

Método de Simpson 1/3 en lugar de usar líneas rectas, usa parábolas para aproximar la curva entre cada grupo de tres puntos. De esta forma, la aproximación se ajusta mejor a funciones que cambian de forma más "curvada", y suele ser más precisa que el método del trapecio para una misma cantidad de intervalos, especialmente cuando la función es suave.

### 12. Calcula la integral aproximada de $f(x) = x^2$ de $x = 0$ a $x = 2$ usando el método del trapecio compuesto con $n=2$ subintervalos. Muestra tu procedimiento.

```python
import numpy as np

def f(x):
    return x**2 

n = 2 # subintervalos

a = 0
b = 2

x = np.linspace(a, b, n + 1)
y = f(x)

I = (b-a) * ( y[0] + 2 * np.sum(y[1:-1]) + y[n]) / (2*n)

print (f'Integral : {I:.4f}')
```
